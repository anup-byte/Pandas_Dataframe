loc and iloc

loc(Label-based Indexing)

df.loc[row_label, column_label]

```python
df = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]}, index = ['x', 'y', 'z'])
```

```python
df
```

```python
df.loc['y', 'A']
```

```python
df.iloc['y', 'A']
```

iloc(Integer-based Indexing)

df.iloc[row_position, column_position]

```python
df
```

```python
df.iloc[1,0]
```

Key differences:

loc uses labels for indexing and iloc uses integer positions.

loc is inclusive for slices(includes the last element), while iloc is exclusive for slices.

loc requires exact labels , whereas iloc works with integer positions regardless of the labels.
