Changing the Index

```python
import pandas as pd
import numpy as np
```

```python
nutrition = pd.read_csv("nutrition.csv", index_col=[0])
```

```python
nutrition.head()
```

```python
nutrition.index
```

```python
type(nutrition.index)
```

```python
pd.RangeIndex(start=0, stop=8789, step=1)
```

```python
nutrition.index = pd.RangeIndex(start=0, stop=8789, step=1)
```

```python
type(nutrition.index)
```

What if we want to change this index from a series of integers to something else?

```python
nutrition.set_index('name')
```

```python
nutrition.set_index('name', inplace=True)
```

```python
nutrition.head()
```

```python
nutrition.set_index('folic_acid', drop=True)
```

```python
nutrition.set_index('folic_acid', drop=False)
```

```python
nutrition.head()
```

```python
nutrition.set_index('folic_acid', drop=False, append=True).head()
```

```python
nutrition.set_index('folic_acid', drop=False, append=True, verify_integrity=False).head()
```

```python
nutrition.calories.value_counts()
```

```python
nutrition.set_index('calories', verify_integrity=True)
```
