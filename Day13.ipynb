Dataframe replace() + A Glimpse at Regex

```python
import pandas as pd
import numpy as np
```

```python
nutrition = pd.read_csv("nutrition.csv", index_col=[0])
```

```python
nutrition.set_index('name', inplace=True)
```

```python
nutrition.head()
```

```python
# "As type"
```

```python
# replace Method -> It replaces one thing with the another
```

```python
nutrition.iloc[:6, :1]
```

```python
nutrition.iloc[:6, :1].info()
```

```python
dfm = nutrition.iloc[:6, :1]
```

```python
dfm.replace(to_replace='100 g', value=100)
```

```python
dfm.replace(to_replace='100 g', value=100).info()
```

```python
dfm.replace('100 g', 100)
```

```python
dfm.replace('g', '')
```

```python
# regex -> regular expressions

# regular expressions is like a sub language that could be used to define text patterns.
```

```python
dfm.replace("\sg", '', regex=True)
```

```python
dfm.replace("\sg", '', regex=True).astype(int)
```

```python
dfm.replace("\sg", '', regex=True).astype(int).info()
```
