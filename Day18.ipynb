Merging Units With Column Names

Day 18

```python
import pandas as pd
import numpy as np
```

```python
nutrition = pd.read_csv("nutrition.csv", index_col=[0])
```

```python
nutrition.set_index('name', inplace=True)
```

```python
nutrition.head()
```

```python
units = nutrition.astype(str).replace('[^a-zA-Z]', '', regex=True)
```

```python
units
```

```python
units.mode()
```

```python
units = units.mode()
```

```python
# Dataframe Iteration
```

```python
for k in units:
    print(k)
```

```python
for k in units:
    print(k, units[k])
```

```python
for k in units:
    print(k, units[k].at[0])
```

```python
for k in units:
    print(units[k].at[0])
```

```python
units = units.replace('', np.nan).dropna(axis=1)
```

```python
units
```

```python
for k in units:
    print(units[k].at[0])
```
